```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def binaryTreePaths(root):
    res = []
    path = []
    def dfs(node, path):
        if not node:
            return
        path.append(node.val)
        if not node.left and not node.right:
            res.append(list(path))
        dfs(node.left, path)
        dfs(node.right, path)
        print(path.pop())
    dfs(root, path)
    return res
```

```python
a = TreeNode(2, None, None)
b = TreeNode(5, None, None)
c = TreeNode(1,a,b)
binaryTreePaths(c)
```

```python
# 创建一个二叉树
a = TreeNode(2, TreeNode(3), TreeNode(4))
b = TreeNode(5, TreeNode(6), TreeNode(7))
root = TreeNode(1, a, b)

# 调用函数并输出结果
paths = binaryTreePaths(root)
print(paths)
```
