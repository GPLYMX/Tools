```python
import os
import cv2
base_path = r'D:\MyCodes\pythonProject\coated_tongue_color\datas\data2all\noval-seg-last-orig'
new_path = r'D:\MyCodes\pythonProject\coated_tongue_color\datas\data2all\noval-seg-last'
# with open(new_path,'w') as f:
for root,dirs,files  in os.walk(base_path):
    for file in files:
        new_files = new_path + "\\" +file.replace('jpg','png')
        old_image = cv2.imread(root + "\\" + file)
        cv2.waitKey(0)
        cv2.imwrite(new_files, old_image)
```
