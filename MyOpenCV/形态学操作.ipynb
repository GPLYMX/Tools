```python
import numpy as np
from matplotlib import pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('t.jpg')
```

```python
plt.imshow(img)
```

# 腐蚀和膨胀

```python
kernel = np.ones((5, 5), np.uint8)
```

```python
img2 = cv.erode(img, kernel)
```

```python
plt.imshow(img2)
```

```python
img3 = cv.dilate(img,kernel)
```

```python
plt.imshow(img3)
```

# 开闭运算

```python
# 开运算先腐蚀后膨胀，消除粉刺；闭运算先膨胀再腐蚀，补充噪点
```

```python
kernel = np.ones((5, 5), np.uint8)
```

```python
img = cv.imread('t.jpg')
```

```python
img4 = cv.erode(img3, kernel)
```

```python
plt.imshow(img4)
```

# 礼帽和黑帽

```python
img5 = img-img3
```

```python
plt.imshow(img5[:,:,::-1])
```
