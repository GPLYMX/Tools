```python
img = cv.imread('t.jpg')
```

```python
rows, cols = img.shape[:2]
```

```python
# 
M = np.float32([[1, 0, 100], [0, 1, 50]])
```

```python
dst = cv.warpAffine(img, M, (cols, rows))
```

```python
plt.imshow(dst[:,:,::-1])
```

```python
# 
M = cv.getRotationMatrix2D([100, 100], 90, 1)
```

```python
dst = cv.warpAffine(img, M, (cols, rows))
```

```python
plt.imshow(dst)
```
