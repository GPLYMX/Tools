```python
img = cv.imread('t.jpg')
```

```python
rows, cols = img.shape[0:2]
```

```python
# rows和cols的位置要交换
res = cv.resize(img, (2*cols, 2*rows))
```

# 图像平移

```python
img = cv.imread('t.jpg')
```

```python
rows, cols = img.shape[:2]
```

```python
# 平移矩阵
M = np.float32([[1, 0, 100], [0, 1, 50]])
```

```python
dst = cv.warpAffine(img, M, (cols, rows))
```

```python
plt.imshow(dst[:,:,::-1])
```

# 图像旋转

```python
img = cv.imread('t.jpg')
```

```python
# 获取旋转矩阵
M = cv.getRotationMatrix2D([100, 100], 90, 1)
```

```python
dst = cv.warpAffine(img, M, (cols, rows))
```

```python
plt.imshow(dst)
```
