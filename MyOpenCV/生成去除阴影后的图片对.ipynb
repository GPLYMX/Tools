```python
def mean_value(arr):
    """ 计算非零元素均值"""
    exist = (arr != 0)
    return arr.sum()/exist.sum()

def cal_feature_point(lst):
    lst = np.array(lst)
    local_max1 = argrelextrema(lst, np.greater)
    # print(local_max1)
    # print(lst[local_max1])
#     local_max1 = np.array(local_max1)[0]
#     local_max1 = local_max1.tolist()
#     print(local_max1)
#     local_max2 = argrelextrema(np.array(local_max1), np.greater, order=1)
#     print(local_max2)
    plt.plot(lst)
    plt.scatter(
        local_max1,
        lst[local_max1],
        c='red'
    )
def wipe_shadow(img):
    lst = []
    for i in range(img.shape[0]):
        # if np.any(img[i,:, :]):
        lst.append(mean_value(img[i,:,:]))
#     lst2 = []
#     for i in range(1, len(lst)):
#         lst2.append(lst[i]-lst[i-1])
    cal_feature_point(lst)
```
