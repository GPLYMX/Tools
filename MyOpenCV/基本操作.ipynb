# 图像的加法与混合

```python
img = cv.imread('T.jpg')
```

```python
x = np.uint8([250])
```

```python
y = np.uint8([10])
```

```python
# cv加法取上限
cv.add(x, y)
```

```python
# numpy加法去模
x + y
```

```python
img2 = img + img
```

```python
plt.imshow(img2)
```

```python
img3 = cv.add(img, img)
```

```python
plt.imshow(img3)
```

```python
# 图像混合
img4 = cv.addWeighted(img2, 0.7, img3, 0.3, 4)
```

```python
plt.imshow(img4)
```
